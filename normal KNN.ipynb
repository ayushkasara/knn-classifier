```python
one_hot_encoded_data = pd.get_dummies(df, columns = ['Clubs and interests in college', 'Undergrad (B.Tech, B.E, B.S) major in', 'Percentage (%) of marks scored in 12th', 'CGPA in college', 'Projects completed before graduation', 'Internships completed before graduation'])
print(one_hot_encoded_data)
```

```python
one_hot_encoded_data.head(20)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(one_hot_encoded_data.drop('Sector/Field of work', axis = 1))
scaled_features = scaler.transform(one_hot_encoded_data.drop('Sector/Field of work', axis = 1))

df_feat = pd.DataFrame(scaled_features, columns = one_hot_encoded_data.columns[:-1])
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
      scaled_features, df, test_size = 0.30)
```
